```python
import re
from collections import defaultdict
dct=defaultdict(set)
def read_file(path,f_name):
    with open('{}/{}'.format(path,f_name),'r',encoding='ISO-8859-1') as f:
        data = f.read()
    return data
```

```python
lst = re.findall('((?:^|\n)[a-zA-Z]+.*?:.*?);',read_file('/home/user','c.g4'),re.DOTALL)
```

```python
resp = [[y.strip() for y in re.split(r'\\t|\||\'',x) if y] for x in lst[1].split('\n')[2:]]
```

```python
id_word=re.findall('(.*?):',lst[1])
id_word = id_word[0]
```

```python

for l in resp:
    if len(l) ==2:
        print(l[1])
        dct[id_word].add(l[1])
```

```python
import nltk
from nltk.tokenize import mwe
stopwords = set(nltk.corpus.stopwords.words('english'))
tokenizer = mwe.MWETokenizer(separator='')
TOKEN_MIN_LENGTH = 3
seen = []
def add_new_token(string):
    tokenizer.add_mwe(r'{}'.format(string))
    seen.append(r'{}'.format(string))
```
